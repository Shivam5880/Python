```python
#Q1

from datetime import datetime

class Employee:
    def __init__(self):
        self.EmployeeID = None
        self.Gender = None
        self.Salary = None
        self.PerformanceRating = None

    def get(self):
        self.EmployeeID = input("Enter Employee ID: ")
        self.Gender = input("Enter Gender: ")
        self.Salary = int(input("Enter Salary: "))
        self.PerformanceRating = int(input("Enter Performance Rating (out of 5): "))

class JoiningDetail:
    def __init__(self):
        self.DateOfJoining = None

    def getDoJ(self):
        date_str = input("Enter Date of Joining (YYYY-MM-DD): ")
        self.DateOfJoining = datetime.strptime(date_str, "%Y-%m-%d")

class Information(Employee, JoiningDetail):
    def __init__(self):
        Employee.__init__(self)
        JoiningDetail.__init__(self)
        self.employees = []

    def readData(self):
        n = int(input("Enter number of employees: "))
        for _ in range(n):
            employee = Information()
            employee.get()
            employee.getDoJ()
            self.employees.append(employee)

        # Sort employees by PerformanceRating and then by DateOfJoining
        self.employees.sort(key=lambda x: (-x.PerformanceRating, x.DateOfJoining))

        # Get top 3 employees based on PerformanceRating
        top_employees = self.employees[:3]

        # Sort top 3 employees by DateOfJoining
        top_employees.sort(key=lambda x: x.DateOfJoining)

        # Display the details of top 3 employees
        for emp in top_employees:
            print(f"\nEmployee ID: {emp.EmployeeID}")
            print(f"Gender: {emp.Gender}")
            print(f"Salary: {emp.Salary}")
            print(f"Performance Rating: {emp.PerformanceRating}")
            print(f"Date Of Joining: {emp.DateOfJoining.strftime('%Y-%m-%d')}")

# Create Information object and read data
info = Information()
info.readData()
```

```python
#Q3

#Highest score for in one single match
def Max_Score(matches):
    max_score = 0
    top_player = ""

    for match in matches.values():
        for player, score in match.items():
            if score > max_score:
                max_score = score
                top_player = player

    return (top_player, max_score)

# Test the function with the given input
input_scores = {
    'test1': {'Dhoni': 56, 'Balaji': 85},
    'test2': {'Dhoni': 87, 'Balaji': 200}
}

result = Max_Score(input_scores)
print(result)  # Output: ('Balaji', 200)
```

```python
# highest Score in both match combined
def Max_Score(matches):
    # Dictionary to store total scores of each player
    total_scores = {}

    # Iterate over each test match
    for match in matches.values():
        # Iterate over each player in the test match
        for player, score in match.items():
            # Add the score to the player's total score
            if player in total_scores:
                total_scores[player] += score
            else:
                total_scores[player] = score

    # Find the player with the highest total score
    top_player = max(total_scores, key=total_scores.get)
    top_score = total_scores[top_player]

    return (top_player, top_score)

# Input scores for the test matches
input_scores = {
    'test1': {'Dhoni': 56, 'Balaji': 94},
    'test2': {'Balaji': 80, 'Dravid': 105}
}

# Get the player with the highest total score
result = Max_Score(input_scores)
print(result)  # Output: ('Balaji', 174)
```

```python
#Q4
import random

def card_game():
    # Create a deck of cards (values from 1 to 13)
    deck = list(range(1, 14)) * 4  # 4 suits, each with 13 cards

    # Randomly select 8 cards from the deck
    selected_cards = random.sample(deck, 8)

    # Initialize score
    score = 0

    # Show the first card
    current_card = selected_cards[0]
    print(f"The first card is: {current_card}")

    #  remaining cards
    for i in range(1, len(selected_cards)):
        next_card = selected_cards[i]
        prediction = input("Will the next card be higher or lower? (Enter 'higher' or 'lower'): ").lower()

        # Check if the prediction is correct
        if (prediction == "higher" and next_card > current_card) or (prediction == "lower" and next_card < current_card):
            print(f"Correct! The next card was: {next_card}")
            score += 20
        else:
            print(f"Incorrect. The next card was: {next_card}")
            score -= 15

        # Update the current card to the next card
        current_card = next_card

    print(f"Game over! Your final score is: {score}")

# Run the card game
card_game()
```

```python
#Q7

from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, color):
        self._color = color  # Protected member
        self.__type = "Generic Shape"  # Private member

    @abstractmethod
    def area(self):
        pass

    def describe(self):
        print(f"This is a {self._Shape__type} of color {self._color}")

class Circle(Shape):
    def __init__(self, radius, color):
        super().__init__(color)
        self._radius = radius  # Protected member
        self.__type = "Circle"  # Override private member

    def area(self):
        return 3.14 * self._radius ** 2

    def describe(self):
        print(f"This is a {self._Shape__type} with radius {self._radius} and color {self._color}")

class Rectangle(Shape):
    def __init__(self, width, height, color):
        super().__init__(color)
        self._width = width  # Protected member
        self._height = height  # Protected member
        self.__type = "Rectangle"  # Override private member

    def area(self):
        return self._width * self._height

    def describe(self):
        print(f"This is a {self._Shape__type} with width {self._width}, height {self._height} and color {self._color}")

# Demonstrate Polymorphism
def shape_info(shape):
    shape.describe()
    print(f"Area: {shape.area()}")

# Creating instances of Circle and Rectangle
circle = Circle(5, "red")
rectangle = Rectangle(4, 6, "blue")

# Using the polymorphic function
shape_info(circle)  # Output: This is a Circle with radius 5 and color red. Area: 78.5
shape_info(rectangle)  # Output: This is a Rectangle with width 4, height 6 and color blue. Area: 24

# Accessing protected and private members directly (not recommended)
print(circle._color)  # Output: red
print(rectangle._color)  # Output: blue

# Accessing private member directly using name mangling (not recommended)
print(circle._Shape__type)  # Output: Circle
print(rectangle._Shape__type)  # Output: Rectangle
```
